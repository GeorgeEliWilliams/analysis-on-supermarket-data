```python
#import your packages here
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the sales data here
sales_data = pd.read_csv('sales_data.csv')
sales_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#Question 1
```

```python
sales_data.head()
```

```python
markets = list(sales_data['Market'].unique())
average_sales = []
average_profits = []
avg_sale_profit_df = pd.DataFrame()

for market in markets:
    market_df = sales_data[sales_data['Market'] == market]
    average_sale = market_df['Sales'].mean()
    average_profit = market_df['Profit'].mean()
    
    average_sales.append(average_sale)
    average_profits.append(average_profit)

avg_sale_profit_df['Market'] = markets
avg_sale_profit_df['Average Sales'] = average_sales
avg_sale_profit_df['Average Profit'] = average_profits 
```

```python
#average sales and profit per market
```

```python
avg_sale_profit_df
```

```python
#a stacked bar chat showing the sales and profit per market
```

```python
plt.bar(avg_sale_profit_df['Market'], avg_sale_profit_df['Average Sales'])
plt.bar(avg_sale_profit_df['Market'], avg_sale_profit_df['Average Profit'], 
        bottom=avg_sale_profit_df['Average Sales'])

plt.legend(['Average Sales', 'Average Profit'])
plt.title('Average Sales & Average Profit By Market')
plt.show()
```

```python
#question 1 iii
```

```python
products = list(sales_data['Product Name'].unique())

total_sales = []
total_sales_df = pd.DataFrame()

for product in products:
    product_df = sales_data[sales_data['Product Name']==product]
    total_sale = product_df['Sales'].sum()
    total_sales.append(total_sale)

total_sales_df['Product Name'] = products
total_sales_df['Total Sales'] = total_sales
```

```python
total_sales_df.head()
```

```python
#question 2. THE BEST 10 PERFORMING PRODUCTS WITH REGARDS TO SALES IN 2013
```

```python
total_sales_df_descending = total_sales_df.sort_values(['Total Sales'], ascending=False)
total_sales_df_descending.head(10)
```

```python
# THE LEAST 10 PERFORMING PRODUCTS WITH REGARDS TO SALES IN 2013
```

```python
total_sales_df_ascending = total_sales_df.sort_values(['Total Sales'], ascending=True)
total_sales_df_ascending.head(10)
```

```python
#QUESTION 3
```

```python
dates = sales_data['Order Date']
days = []
months = []

for date in dates:
    splitted_date = date.split('-')
    
    day = splitted_date[-1]
    days.append(day)
    
    month = splitted_date[-2]
    months.append(month)

sales_data['Days'] = days
sales_data['Months'] = months
```

```python
sales_data.head()
```

```python
# THE LINE PLOT SHOWS THE TREND OF PROFIT FROM JANUARY TO DECEMBER
```

```python
profit_data_df = sales_data[['Order Date','Profit']]

profit_data_df['Order Date'] = pd.to_datetime(profit_data_df['Order Date'])
profit_data_df.set_index('Order Date', inplace=True)0
plt.rcParams["figure.figsize"] = (20,10)
profit_data_df.plot()
```
